```python
import pandas as pd

df = pd.read_csv("../data/tweets.csv")
```

```python
df.info()
```

```python
tweets = df["text"]

tweets.head()
```

## TF-IDF

!["TFIDF"](imgs/tfidf.jpeg)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

VOCAB_SIZE = 1000

vectorizer = TfidfVectorizer(max_features=VOCAB_SIZE)

X = vectorizer.fit_transform(tweets)

X
```

```python
X = X.toarray()
X
```

```python
X.max()
```

```python
X.mean()
```
