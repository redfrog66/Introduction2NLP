### But why?

```python
%%timeit

my_list = list(range(1000000))

my_list = [2 * i for i in my_list]
```

```python
%%timeit

my_array = np.zeros(1000000)

my_array *= 2
```

```python
sizes = 10 ** np.array(range(1, 7+1))
sizes
```

```python
from timeit import timeit

times_list = []
times_numpy = []

for size in sizes:
    _input = list(range(size))

    times_list.append(timeit(lambda: [2 * element for element in _input], number=100))

for size in sizes:
    _input = np.array(range(size))

    times_numpy.append(timeit(lambda: 2 * _input, number=100))
```

```python
plt.plot(sizes, times_list, color="cyan")
plt.plot(sizes, times_numpy, color="orange")
plt.legend(["list", "numpy"], prop={"size": 16})
plt.xlabel("number of elements")
plt.ylabel("seconds")
plt.xscale("log")
plt.show()
```

```python
print(times_list)
print(times_numpy)
print([l/n for l, n in zip(times_list, times_numpy)])
```

### Basic operations

```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print(a - b)
print(a + b)
print(2 * a)
print(a ** 2)
print(a < 2)
print(np.sin(a))
print(np.sqrt(a))
print(np.log2(a))
```

```python
a += b
print(a)

a -= b
print(a)
```

```python
a = np.array([1, 2, 3])
b = a

b[0] = 10

print(a)
print(b)
```

```python
a = np.array([1, 2, 3])
b = a.copy()

b[0] = 10

print(a)
print(b)
```

### Matrix operations

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[1, 2, 3, 4], [4, 3, 2, 1], [0, 1, 2, 3]])

C = np.matmul(A, B)
D = A @ B

print("A:", A.shape)
print("B:", B.shape)
print("2A:\n", 2 * A)
print("C:\n", C)
print("C shape:", C.shape)
print("D:\n", D)
print("D shape:", D.shape)
```
